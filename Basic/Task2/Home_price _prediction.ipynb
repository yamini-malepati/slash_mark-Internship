```python
# Linear Regression and Gradient Boosting
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Prepare data
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date]
data['date'] = conv_dates
train_data = data.drop(['id', 'price'], axis=1)
```

```python
# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(train_data, labels, test_size=0.10, random_state=2)
```

```python
# Linear Regression
reg = LinearRegression()
reg.fit(x_train, y_train)
reg_score = reg.score(x_test, y_test)
```

```python
# Gradient Boosting
clf = GradientBoostingRegressor(n_estimators=400, max_depth=5, min_samples_split=2, learning_rate=0.1, loss='squared_error')
clf.fit(x_train, y_train)
clf_score = clf.score(x_test, y_test)
```

```python
# Plot training scores
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(np.arange(1, 401), clf.train_score_, 'b-', label='Training Set')
plt.title("Training Scores")
plt.xlabel("Iterations")
plt.ylabel("Score")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error

# Plot deviance
t_sc = np.zeros((400,), dtype=np.float64)
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)

test_sc = np.arange(1, 401)
plt.subplot(1, 2, 2)
plt.plot(test_sc, clf.train_score_, 'b-', label='Training Set')
plt.plot(test_sc, t_sc, 'r-', label='Test Set')
plt.title("Deviance")
plt.xlabel("Iterations")
plt.ylabel("Score")
plt.legend()
plt.show()
```
